```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df['flower_names'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
df0 = df[:50]
df0.head()
len(df0)
new_df0 = df0.drop(['target', 'flower_names'], axis =1)
new_df0.head()
```

```python
df1 = df[50:100]
df1.head()
len(df1)
new_df1 = df1.drop(['target', 'flower_names'], axis =1)
new_df1.head()
```

```python
df2 = df[100:]
df2.head()
len(df2)
new_df2 = df2.drop(['target', 'flower_names'], axis =1)
new_df2.head()
```

```python
plt.scatter(new_df0['sepal length (cm)'], new_df0['petal length (cm)'], color = 'red', marker = '*')
plt.scatter(new_df0['sepal width (cm)'], new_df0['petal width (cm)'], color = 'red', marker = '*')
plt.scatter(new_df1['sepal length (cm)'], new_df1['petal length (cm)'], color = 'blue', marker = '*')
plt.scatter(new_df1['sepal width (cm)'], new_df1['petal width (cm)'], color = 'blue', marker = '*')
plt.scatter(new_df2['sepal length (cm)'], new_df2['petal length (cm)'], color = 'green', marker = '*')
plt.scatter(new_df2['sepal width (cm)'], new_df2['petal width (cm)'], color = 'green', marker = '*')
```

```python
input_df = df.drop(['target','flower_names'], axis = 1)
input_df
```

```python
target_df = df['target']
target_df
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(input_df,target_df, test_size = 0.2)
len(X_test)
```

```python
model = SVC()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
model.predict([[5.0,3.6,1.4,0.2]])
```
