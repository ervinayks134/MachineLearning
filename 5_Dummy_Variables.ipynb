```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

```

```python
df = pd.read_csv('homeprices2.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
final = merged.drop(['town'], axis = 'columns')
final
```

```python
final_final = final.drop(['west windsor'], axis = 'columns')
final_final
```

```python
x = final_final.drop(['price'], axis = 'columns')
x
```

```python
y = final.price
```

```python
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
reg.score(x,y)
```

```python

test_data = pd.DataFrame([[4000,False, False]], columns = ['area','monroe township', 'robinsville'])
answer = reg.predict(test_data)
print(f'The predicted price for 4000 sqft. home in west windosr is {answer}')
```
